#### Question
Consider the problem of a single input variable
x and output variable y. Assume we want to fit a simple linear model of the form ŷ =
θ1x + θ0, and accordingly find the best value of θ1 and θ0. Assume we have N
measurements of the input and output where the n-th measurement is given by xn and yn
with corresponding estimate ŷn = θ1xn + θ0. To find the best model, we want to
minimize the mean square error (MSE).

Find the optimum values of θ1 and θ0 that minimizes the MSE.

#### The Normal Equation
To find the value of θ that minimizes the cost function, there is a closed-form solution—in other words, a mathematical equation that gives the result directly. This is called the Normal Equation (Equation 4-4).2

$\hat{\theta}$ = ($X^T$\$X$)$^{-1}$ $X^T$ $y$

where,
* $\hat{\theta}$ is the optimum value of $\theta$ at which the cost function is at its minimum.
* y is the vector of target values containing $y^1$ to $y^m$.


Let’s create some linear-looking data to test this equation and compute and compare the theta_optimised:

```python
# Assumption
N = 1000
theta_0 = 5
theta_1 = 8
```

```python
%matplotlib inline
from matplotlib import pyplot as plt
import numpy as np

X = 2 * np.random.rand(N, 1)
y = theta_0 + theta_1 * X + np.random.randn(N, 1)
```

```python
# our dataset will look like
plt.plot(X, y, "g.")
plt.show()
```

```python
X_b = np.c_[np.ones((N, 1)), X]  # add x0 = 1 to each instance
theta_optimised = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
print("Optimum value of theta_0 is:", theta_optimised[0])
print("Optimum value of theta_1 is:", theta_optimised[1])
```

We would liked to have values of $\theta_0$ = 5 and $\theta_1$ = 8 instead of $\theta_0$ = 5.0312 and $\theta_1$ = 7.9582, still close enough. 

But the noise present in the dataset made it impossible to recover the exact parameters of the original function.
